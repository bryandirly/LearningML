MatPlotLib :

Useful for making plot

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('train.csv')
numerical_cols = df.select_dtypes(include=['float64', 'int64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
print("Numerical columns:", numerical_cols)
print("Categorical columns:", categorical_cols)
```

Creating Plot

```python
x = np.linspace(0, 10, 500)
y = x ** 2
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Parabola')
plt.show()
```

```python
x = np.linspace(0, 100, 500)
y = np.sin(x) * np.exp(-x/100)
plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show()
```

```python
x = np.linspace(0, 100, 500)
y = np.sin(x) * np.exp(-x/100)
plt.plot(x, y, 'g-')
plt.plot(x, np.exp(-x/100), 'r--', alpha=0.8)
plt.axhline(0, color='black', lw=0.5)
plt.grid(True)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show()
```

Bar Plot

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
languages = ['Python', 'Java', 'C++', 'JavaScript']
people = [50, 30, 20, 40]
ax.bar(languages, people, color=['blue', 'orange', 'green', 'red'])
plt.xlabel('Programming Languages')
plt.ylabel('Number of People')
plt.title('Programming Language Popularity')
plt.show()
```

Pie Chart

```python
fig1 = plt.figure()
ax = fig1.add_axes([0, 0, 1, 1])
ax.pie(people, labels=languages, autopct='%1.1f%%', startangle=30)
```

Scatter Plot

```python
x = np.random.rand(50) * 100
y = np.random.rand(50) * 100
plt.scatter(x, y, color='purple', alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Scatter Plot')
```

```python
a = np.linspace(0, 100, 500)
x = np.sin(a)
y = np.exp(-a)
plt.scatter(x, y, color='purple', alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Random Scatter Plot')
```
