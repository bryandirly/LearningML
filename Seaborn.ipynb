Seaborn

- Library for Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

Seaborn has some built-in datasets.

```python
tips = sns.load_dataset("tips")
numerical_features = tips.select_dtypes(include=[np.number]).columns.tolist()
```

```python
tips.head()
```

```python
# Setting a theme for the plot
sns.set_theme(style="whitegrid")
```

```python
# Visualize the data
plt.figure(figsize=(5, 3))
sns.relplot(data=tips, x='total_bill', y='tip', col='time', hue='smoker', style='smoker', size='size')
```

```python
# Load the Iris dataset
iris = sns.load_dataset("iris")
```

```python
iris.head()
```

```python
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=iris)
```

```python
sns.scatterplot(x='sepal_width', y='petal_width', hue='species', data=iris)
```

Count Plot

```python
sns.countplot(x='species', data=iris)
```

Bar Chart

```python
titanic = sns.load_dataset("titanic")
titanic.head()
```

```python
sns.barplot(x='sex', y='survived', hue='class', data=titanic)
```

```python
sns.displot(iris['sepal_length'], kde=True, bins=50)
sns.displot(iris['sepal_length'], kde=True, rug=True)
```

Correlation:
- Positive Correlation
- Negative Correlation
- No Correlation

Heatmap

```python
plt.figure(figsize=(5, 3))
correlation = tips[numerical_features].corr()
sns.heatmap(correlation, cbar=True, square=True, fmt='.2f', annot=True, annot_kws={'size' : 12}, cmap='Blues')
```
